```python
def read():
    with open("unimulti.in") as f:
        stroka = f.readline().strip()
    return stroka

def func(c,x,y):
    rez = 0
    for j in range(len(c)):
        rez += c[j][0]*(x**c[j][1])*(y**c[j][2])
    return rez

def convert(s):
    i=0
    j=0
    c = []
    length = len(stroka)
    while (i < length):
        c.append([1,0,0])
        znak = 1
        if s[i] == "-":
            znak = -1
            i+=1
        elif s[i] =="+":
            i+=1
        if s[i].isdecimal():
            c[j][0] = int(s[i])
            i+=1
        c[j][0] *= znak
        if i == length:
            break
        if s[i]=="x":
            c[j][1] = 1
            i+=1
            if i+1<length:
                if s[i]=="^":
                    i+=1
                    c[j][1] = int(s[i])
                    i+=1
        if i == length:
            break
        if s[i]=="y":
            c[j][2] = 1
            i+=1
            if i+1<length:
                if s[i]=="^":
                    i+=1
                    c[j][2] = int(s[i])
                    i+=1
        j+=1
    return c
        
def out(ch):
    with open("unimulti.out",'w') as f:
        f.write("Multiple local maxima: {}".format(ch[0]))
        f.write("\n")
        f.write("Multiple local minima: {}".format(ch[1]))
        f.write("\n")
        f.write("Plateaus: {}".format(ch[2]))
        
def check_multi(c,rez):
    f = [["z" for i in range(21)]for j in range(21)]
    count = [0 for i in range(3)]
    for y in range(-10, 11):
        for x in range(-10,11):
            y_ind = y+10
            x_ind = x+10
            if f[y_ind][x_ind] == "z":
                f[y_ind][x_ind] = func(c,x,y)
            if x == -10:
                if y == -10:
                    f[y_ind+1][x_ind] = func(c,x,y+1)
                    f[y_ind][x_ind+1] = func(c,x+1,y)
                    if f[y_ind][x_ind]> f[y_ind][x_ind+1] and f[y_ind][x_ind]> f[y_ind+1][x_ind]:
                        count[0]+=1
                        if count[0]>1:
                            rez[0] = "Yes"
                    elif f[y_ind][x_ind]< f[y_ind][x_ind+1] and f[y_ind][x_ind]< f[y_ind+1][x_ind]:
                        count[1]+=1
                        if count[1]>1:
                            rez[1] = "Yes"
                    elif f[y_ind][x_ind]==f[y_ind][x_ind+1] or f[y_ind][x_ind]==f[y_ind+1][x_ind]:
                        rez[2] = "Yes"
                    continue
                elif y == 10:
                    f[y_ind][x_ind+1] = func(c,x+1,y)
                    if f[y_ind][x_ind]> f[y_ind][x_ind+1] and f[y_ind][x_ind]> f[y_ind-1][x_ind]:
                        count[0]+=1
                        if count[0]>1:
                            rez[0] = "Yes"
                    elif f[y_ind][x_ind]< f[y_ind][x_ind+1] and f[y_ind][x_ind]< f[y_ind-1][x_ind]:
                        count[1]+=1
                        if count[1]>1:
                            rez[1] = "Yes"
                    elif f[y_ind][x_ind]==f[y_ind][x_ind+1] or f[y_ind][x_ind]==f[y_ind-1][x_ind]:
                        rez[2] = "Yes"
                    continue
                else:
                    f[y_ind][x_ind+1] = func(c,x+1,y)
                    f[y_ind+1][x_ind] = func(c,x,y+1)
                    if (f[y_ind][x_ind]> f[y_ind][x_ind+1] and f[y_ind][x_ind]> f[y_ind-1][x_ind] and 
                        f[y_ind][x_ind]> f[y_ind+1][x_ind]):
                        count[0]+=1
                        if count[0]>1:
                            rez[0] = "Yes"
                    elif (f[y_ind][x_ind]< f[y_ind][x_ind+1] and f[y_ind][x_ind]< f[y_ind-1][x_ind] and 
                        f[y_ind][x_ind]< f[y_ind+1][x_ind]):
                        count[1]+=1
                        if count[1]>1:
                            rez[1] = "Yes"
                    elif (f[y_ind][x_ind]==f[y_ind][x_ind+1] or f[y_ind][x_ind]==f[y_ind-1][x_ind] or 
                        f[y_ind][x_ind]==f[y_ind+1][x_ind]):
                        rez[2] = "Yes"
                    continue
            elif x == 10:
                if y == -10:
                    f[y_ind+1][x_ind] = func(c,x,y+1)
                    if f[y_ind][x_ind]> f[y_ind][x_ind-1] and f[y_ind][x_ind]> f[y_ind+1][x_ind]:
                        count[0]+=1
                        if count[0]>1:
                            rez[0] = "Yes"
                    elif f[y_ind][x_ind]< f[y_ind][x_ind-1] and f[y_ind][x_ind]< f[y_ind+1][x_ind]:
                        count[1]+=1
                        if count[1]>1:
                            rez[1] = "Yes"
                    elif f[y_ind][x_ind]==f[y_ind][x_ind-1] or f[y_ind][x_ind]==f[y_ind+1][x_ind]:
                        rez[2] = "Yes"
                    continue
                elif y == 10:
                    if f[y_ind][x_ind]> f[y_ind][x_ind-1] and f[y_ind][x_ind] >f[y_ind-1][x_ind]:
                        count[0]+=1
                        if count[0]>1:
                            rez[0] = "Yes"
                    elif f[y_ind][x_ind]< f[y_ind][x_ind-1] and f[y_ind][x_ind] <f[y_ind-1][x_ind]:
                        count[1]+=1
                        if count[1]>1:
                            rez[1] = "Yes"
                    elif f[y_ind][x_ind]==f[y_ind][x_ind-1] or f[y_ind][x_ind]==f[y_ind-1][x_ind]:
                        rez[2] = "Yes"
                    continue
                else:
                    f[y_ind+1][x_ind] = func(c,x,y+1)
                    if (f[y_ind][x_ind]> f[y_ind-1][x_ind] and f[y_ind][x_ind]> f[y_ind][x_ind-1] and 
                        f[y_ind][x_ind]> f[y_ind+1][x_ind]):
                        count[0]+=1
                        if count[0]>1:
                            rez[0] = "Yes"
                    elif (f[y_ind][x_ind]< f[y_ind-1][x_ind] and f[y_ind][x_ind]< f[y_ind][x_ind-1] and 
                        f[y_ind][x_ind]< f[y_ind+1][x_ind]):
                        count[1]+=1
                        if count[1]>1:
                            rez[1] = "Yes"
                    elif (f[y_ind][x_ind]==f[y_ind-1][x_ind] or f[y_ind][x_ind]==f[y_ind][x_ind-1] or 
                        f[y_ind][x_ind]==f[y_ind+1][x_ind]):
                        rez[2] = "Yes"
                    continue
            if y == -10:
                if x!= -10 and x!=10:
                    f[y_ind][x_ind+1] = func(c,x+1,y)
                    f[y_ind+1][x_ind] = func(c,x,y+1)
                    if (f[y_ind][x_ind]> f[y_ind][x_ind+1] and f[y_ind][x_ind]> f[y_ind][x_ind-1] and 
                        f[y_ind][x_ind]> f[y_ind+1][x_ind]):
                        count[0]+=1
                        if count[0]>1:
                            rez[0] = "Yes"
                    elif (f[y_ind][x_ind]< f[y_ind][x_ind+1] and f[y_ind][x_ind]< f[y_ind][x_ind-1] and 
                        f[y_ind][x_ind]< f[y_ind+1][x_ind]):
                        count[1]+=1
                        if count[1]>1:
                            rez[1] = "Yes"
                    elif (f[y_ind][x_ind]==f[y_ind][x_ind+1] or f[y_ind][x_ind]==f[y_ind][x_ind-1] or 
                        f[y_ind][x_ind]==f[y_ind+1][x_ind]):
                        rez[2] = "Yes"
                    continue
            elif y == 10:
                if x!= -10 and x!=10:
                    f[y_ind][x_ind+1] = func(c,x+1,y)
                    if (f[y_ind][x_ind]> f[y_ind][x_ind+1] and 
                        f[y_ind][x_ind]> f[y_ind][x_ind-1] and 
                        f[y_ind][x_ind]> f[y_ind-1][x_ind]):
                        count[0]+=1
                        if count[0]>1:
                            rez[0] = "Yes"
                    elif (f[y_ind][x_ind]< f[y_ind][x_ind+1] and 
                          f[y_ind][x_ind]< f[y_ind][x_ind-1] and 
                          f[y_ind][x_ind]< f[y_ind-1][x_ind]):
                        count[1]+=1
                        if count[1]>1:
                            rez[1] = "Yes"
                    elif (f[y_ind][x_ind]==f[y_ind][x_ind+1] or f[y_ind][x_ind]==f[y_ind][x_ind-1] or 
                        f[y_ind][x_ind]==f[y_ind-1][x_ind]):
                        rez[2] = "Yes"
                    continue
            f[y_ind][x_ind+1] = func(c,x+1,y)
            f[y_ind+1][x_ind] = func(c,x,y+1)
            if (f[y_ind][x_ind]>f[y_ind][x_ind-1] and 
                f[y_ind][x_ind]>f[y_ind][x_ind+1] and 
                f[y_ind][x_ind]>f[y_ind-1][x_ind] and 
                f[y_ind][x_ind]>f[y_ind+1][x_ind]):
                count[0]+=1
                if count[0]>1:
                    rez[0] = "Yes"
            elif (f[y_ind][x_ind]<f[y_ind][x_ind-1] and 
                  f[y_ind][x_ind]<f[y_ind][x_ind+1] and 
                  f[y_ind][x_ind]<f[y_ind-1][x_ind] and 
                  f[y_ind][x_ind]<f[y_ind+1][x_ind]):
                count[1]+=1
                if count[1]>1:
                    rez[1] = "Yes"  
            elif (f[y_ind][x_ind]==f[y_ind][x_ind-1] or f[y_ind][x_ind]==f[y_ind][x_ind+1] or 
                f[y_ind][x_ind]==f[y_ind-1][x_ind] or f[y_ind][x_ind]==f[y_ind+1][x_ind]):
                rez[2] = "Yes"
    return rez
            
if __name__ == "__main__":
    stroka = read()
    coefficents = convert(stroka)
    result = ["No" for i in range(3)]
    result = check_multi(coefficents,result)
    out(result)
```
